```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("911.csv", header=0)
```

```python
dataset.head()
```

```python
#Right now I am only interested in Title, Timestamp, and twp
calls = dataset[['title', 'timeStamp', 'twp']]
```

```python
calls.head()
```

```python
calls['title'].value_counts()
```

```python
calls['twp'].value_counts()
```

```python
vehical_accident_check = calls['title'] == "Traffic: VEHICLE ACCIDENT -"
vehical_accidents = calls[vehical_accident_check]
vehical_accidents.head()
```

```python
vehical_accidents['twp'].value_counts()
```

```python
is_victim = calls['title'] == "EMS: FALL VICTIM"
victims = calls[is_victim]
victims['twp'].value_counts()
```

```python
#top 15 Calls by title
calls['title'].value_counts().head(10)
```

```python
#bar chart of this record above
import matplotlib.pyplot as plt
top10Type = calls['title'].value_counts().head(10)

top10Type.plot.bar()
```

```python
top10Town = calls['twp'].value_counts().head(10)

top10Town.plot.bar()
```

```python
isAccident = calls['title'] == "Traffic: VEHICLE ACCIDENT -"
accidents = calls[isAccident]
top10TownAccidents = accidents['twp'].value_counts().head(10)
top10TownAccidents.plot.bar()
```

```python
#Date /Time analysis
calls['timeStamp'] = pd.to_datetime(calls['timeStamp'])
calls['Year'] = calls['timeStamp'].apply(lambda time: time.year) 
 
```

```python
calls['Year'] = calls['timeStamp'].apply(lambda time: time.year) 
```

```python
calls['Month'] = calls['timeStamp'].apply(lambda time: time.month)
```

```python
calls['Day'] = calls['timeStamp'].apply(lambda time: time.day) 
```

```python
calls['Hour'] = calls['timeStamp'].apply(lambda time: time.hour) 
```

```python
calls.head(20)
```

```python
calls['Year'].value_counts().head(4).plot.bar()
```

```python
calls['Month'].value_counts().head(12).plot.bar()
```

```python
isAccident = calls['title'] == "Traffic: VEHICLE ACCIDENT -"
accidents = calls[isAccident]
accidents['Month'].value_counts().head(12).plot.bar()
```

```python
calls['Hour'].value_counts().head(24).plot.bar()
```

```python
isAccident = calls['title'] == "Traffic: VEHICLE ACCIDENT -"
accidents = calls[isAccident]
accidents['Hour'].value_counts().head(24).plot.bar()
```

```python
isResp = calls['title'] == "EMS: RESPIRATORY EMERGENCY"
resps = calls[isResp]
resps['Hour'].value_counts().head(24).plot.bar()
```

```python
resps['Month'].value_counts().head(24).plot.bar()
```

```python
bymonth = calls.groupby("Month").count()
bymonth
```

```python
pvt = pd.pivot_table(data=bymonth, index="Month", columns="Hour", aggfunc=np.sum)
pvt
```
